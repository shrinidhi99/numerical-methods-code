```python
def two_point_first_der(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>Two point first der')
    print('>>>>Forward')
    for i in range(0,n-1):
        der=(y[i+1]-y[i])/h
        print(f'{x[i]}\t :', der)
    print(f'{x[-1]}\t :', 'Null')
    print('>>>>Backward')
    print(f'{x[0]}\t :', 'Null')
    for i in range(1,n):
        der=(y[i]-y[i-1])/h
        print(f'{x[i]}\t :', der)

def three_point_LEP(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>Three point LEP first der')
    for i in range(0,n-2):
        der=(-3*y[i]+4*y[i+1]-y[i+2])/(2*h)
        print(f'{x[i]}\t :', der)
    for i in range(n-2, n):
        print(f'{x[i]}\t :', 'Null')
        
def three_point_REP(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>Three point REP first der')
    for i in range(0, 2):
        print(f'{x[i]}\t :', 'Null')
    for i in range(2,n):
        der=(-3*y[i]+4*y[i-1]-y[i-2])/(2*-h)
        print(f'{x[i]}\t :', der)

def three_point_midpoint(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>Three point midpoint first der')
    for i in range(0, 1):
        print(f'{x[i]}\t :', 'Null')
    for i in range(1,n-1):
        der=(-y[i-1]+y[i+1])/(2*h)
        print(f'{x[i]}\t :', der)
    for i in range(n-1, n):
        print(f'{x[i]}\t :', 'Null')

def five_point_LEP(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>five point LEP first der')
    for i in range(0,n-4):
        der=(-25*y[i]+48*y[i+1]-36*y[i+2]+16*y[i+3]-3*y[i+4])/(12*h)
        print(f'{x[i]}\t :', der)
    for i in range(n-4, n):
        print(f'{x[i]}\t :', 'Null')

def five_point_REP(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>five point REP first der')
    for i in range(0, 4):
        print(f'{x[i]}\t :', 'Null')
    for i in range(4,n):
        der=(-25*y[i]+48*y[i-1]-36*y[i-2]+16*y[i-3]-3*y[i-4])/(12*-h)
        print(f'{x[i]}\t :', der)

def five_point_midpoint(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>five point midpoint first der')
    for i in range(0, 2):
        print(f'{x[i]}\t :', 'Null')
    for i in range(2,n-2):
        der=(y[i-2]-8*y[i-1]+8*y[i+1]-y[i+2])/(12*h)
        print(f'{x[i]}\t :', der)
    for i in range(n-2, n):
        print(f'{x[i]}\t :', 'Null')

def three_point_midpoint_double_der(x, y):
    """
    x: list of equally spaced values
    y: list of function values evaluated at each x
    """
    n=len(x)
    h=x[1]-x[0]
    print('>>>>Three point midpoint double der')
    for i in range(0, 1):
        print(f'{x[i]}\t :', 'Null')
    for i in range(1,n-1):
        der=(y[i-1]-2*y[i]+y[i+1])/(h**2)
        print(f'{x[i]}\t :', der)
    for i in range(n-1, n):
        print(f'{x[i]}\t :', 'Null')
```
